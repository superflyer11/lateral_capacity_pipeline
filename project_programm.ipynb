```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define project timeline
start_date = datetime(2024, 6, 1)
tasks = {
    "Literature Review and Data Collection": (0, 120),
    "Meshing": (0, 150),
    "Pipeline Setup": (10, 105),
    "Model Implementation (and Development)": (60, 165),
    "Testing": (90, 195),
    "Validation with Experimental Data": (150, 210),
    "Comparison with Empirical Methods": (150, 210),
    "Report Writing": (180, 210)
}

# Plot the Gantt chart
fig, ax = plt.subplots(figsize=(10, 6))

for i, (task, (start, end)) in enumerate(tasks.items()):
    ax.barh(i, end - start, left=start_date + timedelta(days=start), height=0.8, align='center')
    ax.text(start_date + timedelta(days=start + (end - start) / 2), i, task, va='center', ha='center', color='white', fontsize="small")


prelim_report_date = datetime(2024, 9, 30)
ax.axvline(
    prelim_report_date, 
    color='red', 
    linestyle='--', 
    linewidth=2, 
    label='Prelim Report Due'
)

interim_report_date = datetime(2024, 11, 11)
ax.axvline(
    interim_report_date, 
    color='red', 
    linestyle='--', 
    linewidth=2, 
    label='Interim Report Due'
)


final_report_date = datetime(2024, 12, 30)
ax.axvline(
    final_report_date, 
    color='red', 
    linestyle='solid', 
    linewidth=2, 
    label='Final Report Due'
)


# Formatting the chart
ax.xaxis_date()
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax.invert_yaxis()
ax.set_xlabel('Timeline')
ax.set_title('Project Programme')
plt.tight_layout()
ax.legend()

# Show the plot
plt.savefig("/mofem_install/jupyter/thomas/mfront_example_test/project_programme.png")
plt.show()
```
